## The column definition of the below data frame:

* wf_id
> index

* poly_Acres_AutoCalc
> Final number of acres consumed by the fire

* irwin_InitialLatitude
>Initial latitude of wildfire

* irwin_InitialLongitude	
>Initial longitude of wildfire

* rain_acumm
>Amount of rain accumulation for the two weeks prior to the start of the fire

* mean
>’mean’ temperature value (Fahrenheit) of the past two weeks leading to the start of fire

* Count
>Amount of days over 90 degrees Fahrenheit for the 2 weeks prior to the start of the fire
